```python
import pandas as pd
import re

def clean_ingredients(ingredient):
    if isinstance(ingredient, str):
        # Remove numbers, text within parentheses, words like "cup" or "cups", "teaspoon", and characters like "/" and "-"
        cleaned_ingredient = re.sub(r'\d+|(\([^)]*\))|cup(s)?|teaspoons|teaspoon|tablespoons|tablespoon|inch|ml|grams|to taste|kg|liter|litre|to|/|-', '', ingredient).strip()

        # Remove extra white spaces
        cleaned_ingredient = ' '.join(cleaned_ingredient.split())

        return cleaned_ingredient
    else:
        return ingredient

# Read the CSV file
df = pd.read_csv('filtered_file.csv')
df.dropna(inplace=True)
# Apply the cleaning function to the "Ingredients" column
df['Actual Ingredients'] = df['Ingredients'].apply(clean_ingredients)

# Display the DataFrame with the new column
# print(df.head)
# print(df.iloc[0])
# df.iloc[2].to_dict()
df.to_csv('/content/actual_ingredients.csv', index=False)
print(df.columns)
df.iloc[0].to_dict()
```

```python
import pandas as pd
from langdetect import detect
import re
import nltk

from gensim.models import Word2Vec
from nltk.corpus import stopwords


nltk.download('punkt')
nltk.download('stopwords')
```

```python

# Extract the data from the specified column and create a list
data_list = df['Ingredients'].tolist()
print(type(data_list[0]))

print("List of data from the specified column:", data_list[0])
print(len(data_list))

######################################
# # Preprocessing the data

allText = ""
for i in range(0, len(data_list)):
  allText += data_list[i]+' '

text = re.sub(r'\[[0-9]*\]', ' ', allText)
text = re.sub(r'\s+', ' ', text)
text = text.lower()
text = re.sub(r'\d', ' ', text)
text = re.sub(r'\s+', ' ', text)
# Remove characters like parentheses, backslash, forward slash, hyphen, and comma
text = re.sub(r'[\(\)\\/\-,]', ' ', text)

# Preparing the dataset
sentences = nltk.sent_tokenize(text)
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

for i in range(len(sentences)):
  sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]

print("Sentences: ")
print(sentences[0])


# Training the Word2Vec model
model = Word2Vec(sentences, min_count=1)
```

```python
# Finding Word Vectors
vector = model.wv['salt']
print(vector)
```

```python
# Most similar words
similar = model.wv.most_similar('karela')
print(similar)
```

```python
# Most similar words
similar = model.wv.most_similar('karela')
for word, score in similar:
    print(f"{word}: {score}")
```
